# Example 1: Simple Line Chart

Let's start with the basics and work our way through making a line chart. Line charts are used to track changes over periods of time. When smaller changes exist, line charts are better to use than bar charts. 

The syntax for drawing a line chart is 
   >plt.plot(x, y)
   
'plt' alerts Python that we will be using the Matplotlib library.

'plot' tells Python to draw a line chart. The *x* value appears first in the parenthesis followed by the *y* value.

>plt.show()
    
This command is necessary to display the plot in Jupyter Notebook. 

## Matplotlib Frequently Used Commands
- `plt.figure()`: To create a new figure
- `plt.plot()`: Plot y versus x as lines and/or markers
- `plt.xlabel()`: Set the label for the x-axis
- `plt.ylabel()`: Set the label for the y-axis
- `plt.title()`: Set a title for the axes
- `plt.grid()`: Configure the grid lines
- `plt.legend()`: Place a legend on the axes
- `plt.savefig()`: To save the current figure on the disk
- `plt.show()`: Display a figure

## Draw the simplest line chart

Let's start with a simple line chart of the price of Apple stocks each year. 


```python
plt.plot(year, apl_price)

plt.show()
```

## Basic Build and Aesthetics

```python
plt.plot(year, apl_price, 
        year, ms_price)


plt.show()
```

### Adding Axis Labels

```python
plt.plot(year, apl_price, 
        year, ms_price)

plt.xlabel('Year')
plt.ylabel('Stock Price')

plt.show()
```

### Changing the Line Colors and Types

```python
plt.plot(year, apl_price, 'k--', 
        year, ms_price, 'g:')

plt.xlabel('Year')
plt.ylabel('Stock Price')

plt.show()
```

### Adjust the *xy* Axes Ranges

```python
plt.plot(year, apl_price, 'k--', 
        year, ms_price, 'g:')

plt.xlabel('Year')
plt.ylabel('Stock Price')

plt.axis([2013, 2019, 0, 180])

plt.show()
```

### Set a Title for the Line Chart

```python
plt.plot(year, apl_price, 'k--', 
        year, ms_price, 'g:')

plt.xlabel('Year')
plt.ylabel('Stock Price')

plt.axis([2013, 2019, 0, 180])
plt.title('Title')


plt.show()
```

# Example 2: Single Paired Values Scatter Plot

Scatter plots are used to determine relationships between the two different things. The x-axis is used to measure one event (or variable) and the y-axis is used to measure the other. If both variables increase at the same time, they have a positive relationship. If one variable decreases while the other increases, they have a negative relationship. Sometimes the variables don't follow any pattern and have no relationship.

```python
x = [12,13,4,15,16,7,8,16,19,11,3,2,9,13,12] # Establish our x values
y = [9,9,7,7,5,4,4,7,7,3,5,1,9,7,5]# Establish our y values

'''------------------------------------------'''

plt.scatter(x, y)# Draw the scatter chart 

'''------------------------------------------'''

# Some aesthetic elements 
plt.title("Learn Scatter Plot") # Add a title to the figure
plt.axis([0, 17, 0, 10]) # Add Axes limiters
plt.xlabel('horizontal axis') # Add x axis labels
plt.ylabel('vertical axis') # Add y axis labels

plt.show() # Show the figure in the notebook
```

# Example 3: Histogram

Histograms are used to demonstrate statistical distribution of a population of variables. While they are often confused with a bar chart, they are not and are a different data visualization. 

```python
# Import our needed libraries to visualize and produce data
import matplotlib.pyplot as plt #for visualizing
import numpy as np #For generating data
np.random.seed(42) #insures that the same random numbers are generated

#Suppress the warnings we may receive in this example about an update and depreciation
import warnings
warnings.filterwarnings("ignore")

#Set python to print the visual in the Jupyter Notebook
%matplotlib inline 

# Generate random data that we can visualize
x = np.random.normal(size = 1000)

'''-------------------------------------------'''

#Plot the histogram
plt.hist(x, normed=True, bins=10)

'''-------------------------------------------'''

#Add some clarifying features
plt.xlabel('Distribution')
plt.ylabel('Probability')
plt.title('Histogram')

plt.show()
```

# Example 6: Multiple Values Plot

```python
# Import our necessary libraries
from matplotlib import pyplot as plt
from matplotlib import style #Added a new part of matplotlib, used for styling chart

np.random.seed(42) #insures that the same random numbers are generated

#Suppress the warnings we may receive in this example about an update and depreciation
import warnings
warnings.filterwarnings("ignore")

#Set python to print the visual in the Jupyter Notebook
%matplotlib inline 

#Use the 'ggplot' style sheet for the chart
style.use('ggplot')

#Create some data to plot
x = [5,8,10]
y = [12,16,6]

x2 = [6,9,11]
y2 = [6,15,7]

x3 = [6, 8, 10]
y3 = [13, 16, 6]

'''___________________________________________'''

plt.plot(x,y,'g',label='line one', linewidth=3)#Line Chart for the x|y values
plt.plot(x2,y2,'c',label='line two',linewidth=3)#Line Chart for the x2|y2 values
plt.bar(x3, y3, label='Value three')#Bar Chart for the x3|y3 values

'''___________________________________________'''

#Add some clarifying features
plt.title('Title')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.legend() #Add a legend

plt.grid(True,color='k')#Make the grid readable

#Print out the chart to Jupyter Notebook
plt.show()
```

# Example 7: Pie Chart

Pie charts are the bane of data visualization. A pie chart's intent is to show a proportional relationship to the whole of the data set by each variable. The problem with them (besides their overuse) is that they do not represent proportionality well. Stacked bar charts should always be your first choice over pie charts for data visualization. However, if you must make a pie chart as part of a dashboard or analysis, blame it on your boss.

```python
import matplotlib.pyplot as plt
%matplotlib inline


# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]

#Aesthetics to be included in the plot
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0.1, 0.1, 0.1)  # explode 1st slice

'''___________________________________________'''

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

'''__________________________________________'''

#Add a legend
plt.legend()

#Set the axis to place the legend on the chart
plt.axis('equal')

#Print the chart to Jupyter Notebook
plt.show()
```
